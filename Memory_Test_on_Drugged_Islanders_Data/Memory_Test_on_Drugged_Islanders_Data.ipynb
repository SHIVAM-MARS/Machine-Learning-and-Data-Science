# Memory Test on Drugged Islanders Data

Anti-Anxiety Medicine on Novel Islanders grouped by Happy or Sad Memory Priming

### About Dataset

https://www.kaggle.com/datasets/steveahn/memory-test-on-drugged-islanders-data?resource=download

__Context__

An experiment on the effects of anti-anxiety medicine on memory recall when being primed with happy or sad memories. The participants were done on novel Islanders whom mimic real-life humans in response to external factors.

Drugs of interest (known-as) [Dosage 1, 2, 3]:

    A - Alprazolam (Xanax, Long-term) [1mg/3mg/5mg]

    T - Triazolam (Halcion, Short-term) [0.25mg/0.5mg/0.75mg]

    S- Sugar Tablet (Placebo) [1 tab/2tabs/3tabs]

*Dosages follow a 1:1 ratio to ensure validity

*Happy or Sad memories were primed 10 minutes prior to testing

*Participants tested every day for 1 week to mimic addiction

**Inspiration**
+ How does anti-anxiety medicine affect you differently by age?
+ Is there a level of plateauing in effectiveness of anti-anxiety medicine - if so, at what point?
+ Effect of anti-anxiety medicine on memory recall?
+ Effectiveness of placebos in a test environment?

### Distribution by Age Categories

```python
df['age'].plot(kind='hist')
```

Categorising Age by following factors:

    age < 35 :- young
    35 <= age <= 55 :- Adult
    age > 55 :- Senior

```python
def age_to_cat(x):
  if x > 55:
    value = 'senior'
  elif x < 35:
    value = 'young'
  else:
    value = 'adult'
  return value
```

Create a column age_cat by calling the above function

```python
df['age_cat'] = df['age'].apply(age_to_cat)
df['age_cat']
```

Distribution by age category:

```python
for value in df.columns:
  if df[value].dtype == 'object':
    sb.countplot(data=df, x= value, palette=palette, hue='age_cat')
  else:
    sb.histplot(data=df, x= value,  hue='age_cat', kde=True, alpha=.6)
  plt.title(value, fontsize=18, fontweight='bold')
  plt.xticks(rotation=65)
  plt.show()
```

### Encoding of Categories

#### Happy_Sad_group Encoding

    Happy :- 1
    Sad :- 0

```python
def hp_grp_to_int(x):
  if x == 'H':
    value = 1
  elif x == 'S':
    value = 0
  return value
```

Creating a column name "Happy_Sad_group_int" by calling above function

```python
df['Happy_Sad_group_int'] = df['Happy_Sad_group'].apply(hp_grp_to_int)
# Cross Verification
df['Happy_Sad_group_int'].value_counts() 
```

#### Drug Encoding

    A :- 1
    T :- 0
    S :- 0

```python
def drg_to_int(x):
  if x == 'A':
    value = 1
  elif x == 'T':
    value = 0
  elif x == 'S':
    value = 0
  return value
```

Creating a column name 'Drug_int' by calling above function 

```python
df['Drug_int'] = df['Drug'].apply(drg_to_int)
df['Drug_int'].value_counts()
```
